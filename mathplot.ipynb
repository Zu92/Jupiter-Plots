```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,10,20)
y=x*x
z=x+y
```

```python
plt.plot(x,y,x,x,x,z)
plt.xlabel("x-asis")
plt.ylabel("y-axis")
```

```python
plt.subplot(3,2,1)
plt.plot(x,x, label="el 1")
plt.subplot(3,2,2)
plt.plot(x,y)
plt.subplot(3,2,3)
plt.plot(x,z)
plt.subplot(3,2,4)
plt.plot(y,z)
plt.subplot(3,2,5)
plt.plot(z,y)
plt.subplot(3,2,6)
plt.plot(x,x)
plt.tight_layout()
plt.legend()
```
